# **Corner Edge Detection**

*   **Harris Corner Detection**






```python
#Importing Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Loading Image
flat_chess = cv2.imread('/content/sample_data/Sample Image/800px_COLOURBOX27025640.jpg')
#Converting Image into Red, Green, Blue Color 
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)
```

```python
#Displaying Image
plt.imshow(flat_chess)
```

```python
#In Gray Scale Version
gray_flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_flat_chess,cmap='gray')
```

```python
#Real Chess Board Image
real_chess = cv2.imread('/content/sample_data/Sample Image/Chess_board_opening_staunton.jpg')
real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(real_chess,cmap='gray')
```

```python
gray_real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_real_chess,cmap='gray')
```

Rightnow they're all Integers.

We are going to be Floating point values for the Corner Harris Detection Algotithm.

```python
gray = np.float32(gray_flat_chess)
```

```python
gray
```

**Now time for  -  Corner Harris Detection**




```python
#src - Source Image (Gray scale image)
#blockSize = Neighbourhood size that detect eigenvalues and vectors
#ksize = Aperture parameter for sobell parameter
#k - Harris detector free parameter(0.04 = Default value)

dst = cv2.cornerHarris(src = gray, blockSize=2, ksize = 3, k= 0.04)
```

```python
#Dialation
dst = cv2.dilate(dst, None)
```

```python
#Visualization
flat_chess[dst>0.01*dst.max()] = [255,25,102]
```

```python
plt.imshow(flat_chess)
```

**Now On Real Chess Image**

```python
gray = np.float32(gray_real_chess)
dst = cv2.cornerHarris(src = gray, blockSize=2, ksize = 3, k= 0.04)
dst = cv2.dilate(dst, None)
real_chess[dst>0.01*dst.max()] = [255,25,102]
plt.imshow(real_chess)
```
